```python
import pandas as pd


df = pd.read_csv('sales_test.csv')
df
```

```python
df["Subtotal"] = df.Preco * df.Quantidade
df
```

```python
subtotal = df.Subtotal
receita_total= sum(subtotal.values.tolist())
receita_total
```

```python
df.groupby(df['Nome_Produto']).sum()
```

```python
produtos_selecionados = df[['Nome_Produto', 'Quantidade']]
produtos_selecionados
```

```python
produtosAgrupados = produtos_selecionados.groupby(produtos_selecionados['Nome_Produto'], as_index= False).sum()
produtosAgrupados
```

```python
maior_venda_id = produtosAgrupados.Quantidade.idxmax()

maior_venda_id
```

```python
maior_venda = produtosAgrupados.loc[maior_venda_id, "Nome_Produto"]
maior_venda
```

```python
clientes_subtotal = df[['ID_Cliente', 'Subtotal']]

clientes_subtotal_agrupados = clientes_subtotal.groupby(['ID_Cliente'], as_index=False).sum()

clientes_subtotal_agrupados
```

```python
melhor_cliente = clientes_subtotal_agrupados['Subtotal'].idxmax()
melhor_cliente
```

```python
melhor_cliente_id = clientes_subtotal_agrupados.loc[melhor_cliente, 'ID_Cliente'].item()
melhor_cliente_id
```
